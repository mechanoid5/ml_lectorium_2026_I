```python
from sklearn.datasets import make_blobs

x_blobs, target_blobs = make_blobs(n_samples=1024, centers=4, cluster_std=rng.uniform(size=4)+.7 )
display( x_blobs.shape )
```

```python
from sklearn.datasets import make_moons
x_moons, target_moons = make_moons(n_samples=1024, noise=.09)
display( x_moons.shape )
```

```python
from sklearn.datasets import  make_circles
x_circ, target_circ = make_circles(n_samples=1024, noise=.09)
display( x_circ.shape )
```

```python
from sklearn.datasets import make_regression

x_reg, y_reg = make_regression( n_samples=1024, n_features=1, noise=25., )
display( x_reg.shape, x_reg.shape )
```

```python
fig,ax = plt.subplots(2,2,figsize=(9,7))

ax[0,0].scatter(x_moons[:,0],x_moons[:,1],c=target_moons,s=1.2,alpha=0.5,cmap='rainbow')
ax[0,0].set_title('make_moons')
ax[0,0].grid()

ax[0,1].scatter(x_blobs[:,0],x_blobs[:,1],c=target_blobs,s=1.2,alpha=0.5,cmap='rainbow')
ax[0,1].set_title('make_blobs')
ax[0,1].grid()

ax[1,0].scatter(x_circ[:,0],x_circ[:,1],c=target_circ,s=1.2,alpha=0.5,cmap='rainbow')
ax[1,0].set_title('make_circles')
ax[1,0].grid()

ax[1,1].scatter(x_reg, y_reg,s=1.2,alpha=0.5)
ax[1,1].set_title('make_regression')
ax[1,1].grid()

plt.tight_layout()
```

```python
from sklearn.datasets import make_classification
x_class, target_class = make_classification(n_samples=1024, n_features=27, n_classes=2)
display(x_class.shape)

from sklearn.decomposition import PCA

x_2_class = PCA(n_components=2).fit_transform(x_class)

fig = plt.figure(figsize=(4,3))
plt.scatter(x_2_class[:,0],x_2_class[:,1],c=target_class,s=1.2,alpha=0.5,cmap='rainbow')
plt.title('make_classification')
plt.grid()
plt.show()
```

# работа с файлами

```python
# создадим файлы с данными
n_samples = 1024
df = pd.DataFrame({
     'Name': [ fake.name() for _ in range(n_samples)], 
     'Address': [ fake.address() for _ in range(n_samples)], 
     'Email': [ fake.email() for _ in range(n_samples)], 
     'Phone Number': [ fake.phone_number() for _ in range(n_samples)],
 })

display(df.sample(5))
```

```python
# !mkdir tmp
```

```python
# формат pickle
df.to_pickle('tmp/fake-data.pkl')
pd.read_pickle('tmp/fake-data.pkl')
```

```python
# формат CSV
df.to_csv('tmp/fake-data.csv',sep='\t',index=False)
pd.read_csv('tmp/fake-data.csv',sep='\t')
```

```python
# формат JSON
df.to_json('tmp/fake-data.json',orient='records', force_ascii=False)
pd.read_json('tmp/fake-data.json')
```

```python
# pip install openpyxl
```

```python
# формат MS Excel
df.to_excel('tmp/fake-data.xlsx',index=False)
pd.read_excel('tmp/fake-data.xlsx')
```

```python
# числовые данные
x = rng.normal(-1.,+1.,(1024,256))
y = rng.normal(-1.,+1.,(1024,128))
x.shape,y.shape
```

```python
# формат Numpy
np.save('tmp/fake-data.npy',x)
np.load('tmp/fake-data.npy')
```

```python
# пакуем несколько матриц в формат Numpy
np.save('tmp/fake-data.npy',{'x':x,'y':y})
np.load('tmp/fake-data.npy', allow_pickle=True)
```

```python
# пакуем матрицу в CSV
np.savetxt('tmp/fake-data.tsv',x, delimiter='\t',fmt='%.2f')
np.loadtxt('tmp/fake-data.tsv')
```
