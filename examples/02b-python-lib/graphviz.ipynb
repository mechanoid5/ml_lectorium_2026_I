```python
import graphviz
```

```python
dot = graphviz.Digraph()
dot.node( name='root', )
dot.node( name='node 1', )
dot.edge( tail_name='root', head_name='node 1')
display(dot)
```

```python
dot = graphviz.Digraph()

dot.node( name='r',  label='Root',   )
dot.node( name='n1', label='node 1', )
dot.node( name='n2', label='node 2', ) 
dot.node( name='n3', label='node 3', ) 

dot.edge( tail_name='r', head_name='n1', label='link')
dot.edge( tail_name='r', head_name='n2', )
dot.edge(tail_name='n2', head_name='n3', )
dot.edge(tail_name='n3', head_name='r', label='back link')

display(dot)
```

```python
dot = graphviz.Digraph()

dot.node( name='r',  label='Root',   color='yellow', style='filled' )
dot.node( name='n1', label='node 1', color='lightskyblue', style='filled' )
dot.node( name='n2', label='node 2', color='lightcoral', style='filled' ) 
dot.node( name='n3', label='node 3', color='gray',) 

dot.edge( tail_name='r',  head_name='n1', color='green', label='link', fontcolor='green' )
dot.edge( tail_name='r',  head_name='n2', color='red' )
dot.edge( tail_name='n2', head_name='n3', color='orange' )
dot.edge( tail_name='n3', head_name='r',  color='magenta', label='back link', fontcolor='magenta' )

display(dot)
```
