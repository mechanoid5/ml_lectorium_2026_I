```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from sympy.abc import x,y,w,o,u
from sympy import *
```

```python
# определим выражение
expr = log(1+exp(-x))
display( expr )
```

```python
# возьмём производную и упростим результат
expr.diff(x).simplify()
```

```python
# подставим цифры в выражение, посчитаем результат
arg = np.linspace(-10,10,100)
values = lambdify(x, expr)(arg)
dvalues = lambdify(x,expr.diff(x).simplify())(arg)
```

```python
arg = np.linspace(-10,10,100)
fig,ax = plt.subplots(1,2,figsize=(12,5))
ax[0].plot(arg,values)
ax[0].grid()
ax[1].plot(arg,dvalues)
ax[1].grid()
plt.show()
```
